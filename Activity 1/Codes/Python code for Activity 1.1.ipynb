```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
```

```python
N = 400
x = np.linspace(-2,2,num = N)
y = x
X,Y = np.meshgrid(x,y)
R = (np.sin(2*np.pi*4*X) + 1)*(256/2)
A = np.zeros(np.shape(R))
plt.set_cmap("gray")
plt.imshow(R, cmap = "gray")
plt.colorbar()
```

```python

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)

ax.view_init(-160,80)
ax.plot_surface(X,Y,R)
```

```python
N = 200
x = np.linspace(-1,1,num = N)
y = x
X,Y = np.meshgrid(x,y)
R = np.sqrt(X**2 + Y**2)
A = np.zeros(np.shape(R))
A[np.where(R<0.5)] = 1.0
A[np.where(R<0.1)] = 0.0
plt.imshow(A, cmap = "gray")
```
